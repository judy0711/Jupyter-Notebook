```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### Basic Graph

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# resize your graph(dpi specifies pixels per inch)
plt.figure(figsize=(5,3), dpi = 150)

# Line 1 

# Keyword Argument Notation
#plt.plot(x,y, color = 'red', label = '2x', linewidth = 2, marker = '.', markersize = 10, markeredgecolor = 'black' )


# Shorthand Notation
#fmt = '[color][marker][line]'
plt.plot(x,y,'b^--', label = '2x')

# Line 2

# Select Intervals we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2, x2**2, 'r', label = 'x^2')

# Add a title(specify font parameters with fontdict)
plt.title('Our first graph!', fontdict = {'fontname':'Serif', 'fontsize':20})

# X and Y labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# X, Y axis Tickmarks(scale of your graph)
plt.xticks([0,1,2,3,4])     
#plt.yticks([0,2,4,6,8,10])

# Label the line
plt.legend()    

# Save figure(dpi = 300 is good)
plt.savefig('mygraph.png', dpi = 300)  

# Show Plot
plt.show()
```

```python
# subplots
months = range(12)
temperature = [36, 36, 39, 52, 61, 72, 77, 75, 68, 57, 48, 48]
flights_to_hawaii = [1200, 1300, 1100, 1450, 850, 750, 400, 450, 400, 860, 990, 1000]

plt.subplot(1, 2, 1)
plt.plot(months, temperature)

plt.subplot(1, 2, 2)
plt.plot(temperature, flights_to_hawaii, "o")

plt.show()
```

```python
x = range(7)
straight_line = [0, 1, 2, 3, 4, 5, 6]
parabola = [0, 1, 4, 9, 16, 25, 36]
cubic = [0, 1, 8, 27, 64, 125, 216]

plt.subplot(2,1,1)
plt.plot(x, straight_line)

plt.subplot(2,2,3)
plt.plot(x,parabola)

plt.subplot(2,2,4)
plt.plot(x, cubic)

plt.subplots_adjust(wspace=0.35, bottom = 0.2)
plt.show()
```

```python
# Create legend

months = range(12)
hyrule = [63, 65, 68, 70, 72, 72, 73, 74, 71, 70, 68, 64]
kakariko = [52, 52, 53, 68, 73, 74, 74, 76, 71, 62, 58, 54]
gerudo = [98, 99, 99, 100, 99, 100, 98, 101, 101, 97, 98, 99]

plt.plot(months, hyrule)
plt.plot(months, kakariko)
plt.plot(months, gerudo)

#create your legend here
legend_labels = ["Hyrule", "Kakariko", "Gerudo Valley"]
plt.legend(legend_labels, loc=8)

plt.show()
```

```python
# Set Ticks

month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep","Oct", "Nov", "Dec"]

months = range(12)
conversion = [0.05, 0.08, 0.18, 0.28, 0.4, 0.66, 0.74, 0.78, 0.8, 0.81, 0.85, 0.85]

plt.xlabel("Months")
plt.ylabel("Conversion")

plt.plot(months, conversion)

# Your work here
ax = plt.subplot()
ax.set_xticks(months)
ax.set_xticklabels(month_names)
ax.set_yticks([0.10, 0.25, 0.5, 0.75])
ax.set_yticklabels(["10%", "25%", "50%", "75%"])

plt.show()
```
